```python
%load_ext autoreload
%autoreload 2

import numpy as np
import matplotlib.pyplot as plt
import networkx as nx

from dcglearn import *
```

# Example 1

```python
n_vars = 4
edges = [(0,1),(1,2),(2,1),(3,2)]
names = {i:"$X_%d$"%(i+1) for i in range(n_vars)}

B_support = edges2array(edges, 4) - np.eye(4)

# find equivalence class of true graph
candidate = AdjacencyStucture(n_vars, edge_list=edges)
target = AdjacencyStucture(n_vars, edge_list=[])
search = GraphEquivalenceSearch(candidate, target)

search.search_dfs()
search.expand_column_permutations()
equiv_class = [binary2array(bstr) for bstr in search.visited_graphs]

# plot structure
plot_structure(edges, n_vars, save_path=None, figsize=(2.5,2.5), name_list=names, latex=True, 
               node_size=800, font_size=16, width=2.5, graphviz=False, connectionstyle='arc3,rad=0.15', node_color='skyblue')

plot_collection({"Graph %d"%(i+1):array2edges(g) for i,g in enumerate(equiv_class)}, 4, n_cols=4, name_list=names, node_size=600, font_size=14, width=2, graphviz=False, latex=True, connectionstyle='arc3,rad=0.15', node_color='skyblue')
```
